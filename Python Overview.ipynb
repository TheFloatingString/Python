# Classes

```python
import Coffee

class CupOfCoffee:
    
    def __init__(self, quantity, quality, country):
        self.quantity = quantity
        self.quality = quality
        self.country = country
        self.is_safe = True
    
    def coffeeType(self):
        is_safe = True

        if self.country.lower() == 'antarctica':
            print("DO NOT CONSUME THIS COFFEE!!!")
            is_safe = False

        if self.quality.lower() == 'ketchup':
            print("MAXINE HAS WARNED YOU NOT TO DRINK THIS COFFEE....")
            print("But lAURENCE WOULD DRINK IT :D XD")
            is_safe = False

        for i in range(200):
            print("I LOVE COFFEE!!!")
            
        self.is_safe = is_safe
            
        return is_safe
    
    def printMessage(self):
        if self.is_safe:
            print("Maxine suggests that you drink this because it is from {} !".format(self.country))
        else:
            print("LAURENCE :) (this object is from {})".format(self.country))
```

```python
Francis = CupOfCoffee(250, 'relish', 'Mars')
Bob = CupOfCoffee(10, 'ketchup', 'North Korea aka Pluto')
```

```python
Francis.coffeeType()
Bob.coffeeType()
```

```python
Francis.printMessage()
```
